```python
import matplotlib.pyplot as plt

# Assuming 'job_counts' is a pandas Series or a list
job_counts = pd.Series([10, 20, 15, 5], index=['Data Scientist', 'Software Engineer', 'Data Analyst', 'Business Analyst'])

job_counts = job_counts.sort_values(ascending = True)
plt.barh(job_counts.index, job_counts)
plt.xlabel("Number of Jobs")
plt.ylabel('Count of  Job Postings')
plt.xticks(rotation = 45, ha='right')
plt.title("Postings by Job Counts")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi/2, np.pi/2, 31)
y = np.cos(x)**3

# 1) Remove points where y > 0.7 (using boolean indexing)
mask = y <= 0.7
x1 = x[mask]  # Select elements from x where mask is True
y1 = y[mask]  # Select elements from y where mask is True

# 2) Mask points where y > 0.7 (using numpy.ma)
mask = y > 0.7
y2 = np.ma.masked_where(mask, y)

# 3) Set to NaN where y > 0.7
y3 = y.copy()
y3[mask] = np.nan

plt.plot(x * 0.1, y, 'o-', color='lightgrey', label='Original data')
plt.plot(x1 * 0.4, y1, 'o-', label='Points removed')
plt.plot(x * 0.7, y2, 'o-', label='Masked values')
plt.plot(x * 1.0, y3, 'o-', label='NaN values')
plt.legend()
plt.title('Masked and NaN data')
plt.show()
```
