#Median Salary Vs  Data Science Jobs

```python
# Install datasets Library (if not already installed)
!pip install datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter  # Import FuncFormatter

# Loading Data
try:
    dataset = load_dataset('lukebarousse/data_jobs')
    df = dataset['train'].to_pandas()
except Exception as e:
    print(f"Error loading dataset: {e}")
    df = pd.DataFrame()  # Create an empty DataFrame in case of error

# Data Cleanup (check for empty DataFrame before processing)
if not df.empty:
    df.dropna(subset=['salary_year_avg'], inplace=True)  # Drop rows with NaN in-place

# Define the formatter function
def to_k(x, pos):
    """Convert number to string with 'K' as thousands."""
    return f'${int(x/100)}K'

# Create Formatter using FuncFormatter
formatter = FuncFormatter(to_k)

# Create histogram (check for empty DataFrame before plotting)
if not df.empty:
    plt.figure(figsize=(10, 5))
    plt.hist(df['salary_year_avg'], bins=30, edgecolor='black')

    # Set title and labels
    plt.title('Distribution of Average Yearly Salaries')
    plt.xlabel('Average Yearly Salary')
    plt.ylabel('Number of jobs')

    # Apply the formatter to the x-axis
    plt.gca().xaxis.set_major_formatter(formatter)

    # Show the plot
    plt.show()
```
