```python
# Install datasets Library (if not already Installed)
!pip install datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

#Bar Chart - Counts of Job Postings

```python
job_counts = df.job_title_short.value_counts()
```

```python
job_counts
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.title('Postings by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
job_counts.plot(kind = 'bar')
plt.ylabel('Count of Postings')
plt.xlabel('Count of Postings')
plt.title('Job Postings by Job Title')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
job_counts.plot(kind = 'line')
```
